# **Control Flow â€“ The Complete Guide**

---

## ðŸ”¹ What is Control Flow?

Control flow is **the order in which statements are executed** in a Python program.
By default, Python executes code **line by line, from top to bottom**.
Control flow statements **alter** that order â€” allowing decisions, loops, and branching.

---

# âš–ï¸ 1. Conditional Statements (`if`, `elif`, `else`)

Conditional statements let your code **make decisions** based on whether a condition is `True` or `False`.

### ðŸ§  Syntax:

```python
if condition1:
    # Code block 1
elif condition2:
    # Code block 2
else:
    # Code block 3
```

### âœ… Example:

```python
age = 18

if age < 13:
    print("Child")
elif age < 20:
    print("Teenager")
else:
    print("Adult")
```

**Output:**

```
Teenager
```

### ðŸ’¡ Notes:

* Indentation is **required** in Python.
* Conditions can use:

  * Comparison operators: `<`, `>`, `<=`, `>=`, `==`, `!=`
  * Logical operators: `and`, `or`, `not`
  * Membership operators: `in`, `not in`
  * Identity operators: `is`, `is not`

---

# ðŸ” 2. Loops

Loops allow repetition of code blocks until a condition is met.

---

## ðŸ§© a) `for` Loop

Used to iterate over **iterables** â€” lists, strings, tuples, ranges, etc.

### Example:

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

**Output:**

```
apple
banana
cherry
```

### Example with `range()`:

```python

    print(i)
```

**Output:**

```
1
2
3
4
5
```

---

## ðŸ”„ b) `while` Loop

Executes repeatedly **while** a condition remains `True`.

### Example:

```python

while count < 3:
    print("Hi")
    count += 1
```

**Output:**

```
Hi
Hi
Hi
```

### âš ï¸ Infinite Loop Example:

```python

    print("This will run forever!")
```

Use `break` to stop it.

---

# ðŸ§± 3. Loop Control Statements

These statements change how loops behave.

---

### ðŸ”¸ `break`

Terminates the loop immediately.

```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

**Output:**

```
0
1
2
```

---

### ðŸ”¸ `continue`

Skips the current iteration and continues to the next.

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

**Output:**

```
0
1
3
4
```

---

### ðŸ”¸ `pass`

Does nothing â€” used as a placeholder.

```python

    pass  # Placeholder for future code
```

---

# ðŸ”‚ 4. `else` with Loops

Python lets you use `else` with loops â€” the `else` block runs only if the loop **was not broken** by a `break` statement.

### Example:

```python

    print(i)
else:
    print("Loop completed successfully")
```

**Output:**

```

1
2
3
4

```

### Example with `break`:

```python

    if i == 3:

    print(i)

    print("Loop completed successfully")
```

**Output:**

```

```

âž¡ï¸ The `else` did *not* execute because of the `break`.

# ðŸ§® 5. Nested Control Flow

You can nest `if`, `for`, and `while` statements.

```python

        print(f"{i} is even")

        print(f"{i} is odd")

```

---

# ðŸ§© 6. Comprehensions (Compact Control Flow)

List, dictionary, and set comprehensions allow control flow inside expressions.

### Example: List Comprehension with Condition

```python

```

**Output:**

```

```

---

# âš™ï¸ 7. `match-case` (Structural Pattern Matching â€“ Python 3.10+)

A powerful feature for matching values (like `switch` in other languages).

### Example:

```python
command = "start"

match command:
    case "start":
        print("Starting...")
    case "stop":
        print("Stopping...")
    case _:
        print("Unknown command")
```

**Output:**

```
Starting...
```

### Example with Patterns:

```python
point = (3, 4)

match point:
    case (0, 0):
        print("Origin")
    case (x, 0):
        print(f"X-axis at {x}")
    case (0, y):
        print(f"Y-axis at {y}")
    case (x, y):
        print(f"Point at ({x}, {y})")
```

**Output:**

```
Point at (3, 4)
```

---

# âš¡ 8. Exception Handling as Control Flow

Although primarily for error handling, `try/except` can control flow when errors occur.

### Example:

```python
try:
    value = int("abc")
except ValueError:
    print("Invalid input!")
else:
    print("Conversion successful")
finally:
    print("Done.")
```

**Output:**

```
Invalid input!
Done.
```

---

# ðŸ§  9. Best Practices

âœ… Use clear and meaningful conditions.
âœ… Avoid deep nesting â€” refactor with functions or returns.
âœ… Prefer `for` loops over `while` loops when possible.
âœ… Use comprehensions for concise looping logic.
âœ… Always ensure `while` loops have a clear termination condition.
âœ… Use `match-case` for multiple fixed comparisons.

---

# **Namespaces in Python â€“ Complete Guide**

---

## ðŸ”¹ What is a Namespace?

A **namespace** is a **container (mapping)** that holds **names (identifiers)** and the **objects (values)** they refer to.

In simple terms:

> A **namespace** is like a dictionary where keys are variable names and values are the objects those names point to.

Example:

```python
x = 10
```

Here:

* `x` â†’ **name (identifier)**
* `10` â†’ **object (value)**
* Together, they live inside a **namespace**.

---

## ðŸ§± 1. Why Namespaces Exist

Namespaces prevent **name conflicts** in large programs.
Multiple parts of your code can use the same variable name without interfering with each other, as long as they exist in **different namespaces**.

---

## ðŸ§­ 2. Types of Namespaces in Python

Python has **four main types** of namespaces, depending on where and how names are defined:

| Namespace Type          | Created When                | Scope                               |
| ----------------------- | --------------------------- | ----------------------------------- |
| **Built-in Namespace**  | Python interpreter starts   | Available everywhere                |
| **Global Namespace**    | A module (file) is loaded   | Available within the module         |
| **Enclosing Namespace** | Inside nested functions     | Available to inner functions        |
| **Local Namespace**     | Inside a function or method | Available only inside that function |

Letâ€™s go through each one ðŸ‘‡

---

## ðŸ§© 3. Built-in Namespace

Contains all **built-in functions, exceptions, and constants**.

Examples:

```python
print(len)
print(type)
print(range)
print(id)
```

You can view all built-ins:

```python
print(dir(__builtins__))
```

âœ… **Always available** â€” you donâ€™t need to import anything.

---

## ðŸŒ 4. Global Namespace

Created when a Python file (module) is executed.
Contains names defined at the **top level** of the module.

### Example:

```python
x = 100  # Global variable

def show():
    print(x)  # Accessible inside the function

show()
```

**Output:**

```
100
```

---

## ðŸ”’ 5. Local Namespace

Created when a **function is called**.
Contains **names defined inside** the function.

### Example:

```python

    y = 5  # Local variable
    print(y)

func()
# print(y)  # âŒ Error: y not defined
```

**Output:**

```

```

ðŸ§  When the function finishes, its local namespace is **destroyed**.

---

## ðŸª† 6. Enclosing Namespace (Nested Functions)

Appears in **nested functions** â€” inner functions can access variables from the outer function.

### Example:

```python

    x = "outer variable"

    def inner():
        print(x)  # Access enclosing variable

    inner()

outer()
```

**Output:**

```

```

---

## ðŸ§® 7. LEGB Rule (Name Resolution Order)

When you reference a name, Python looks for it in this order:

> **L â†’ E â†’ G â†’ B**

| Order             | Namespace                 | Example                                   |
| ----------------- | ------------------------- | ----------------------------------------- |
| **L (Local)**     | Inside current function   | local variables                           |
| **E (Enclosing)** | Inside enclosing function | variables in outer (non-global) functions |
| **G (Global)**    | Module-level variables    | variables defined at top level            |
| **B (Built-in)**  | Python built-in names     | `len`, `print`, etc.                      |

### Example:

```python
x = "global"

def outer():
    x = "enclosing"

    def inner():
        x = "local"
        print(x)  # Looks in local â†’ enclosing â†’ global â†’ built-in

    inner()

outer()
```

**Output:**

```
local
```

---

## ðŸ§  8. The `globals()` and `locals()` Functions

You can inspect or modify namespaces dynamically.

### `globals()` â†’ returns the global namespace (as a dictionary)

```python

```

### `locals()` â†’ returns the local namespace (inside functions)

```python

    y = 5
    print(locals())

```

**Output:**

```

```

---

## âš™ï¸ 9. The `global` and `nonlocal` Keywords

Used to modify variables in **outer scopes**.

---

### ðŸ”¸ `global` â€“ access and modify global variables

```python

    global x
    x = 10

```

**Output:**

```

```

---

### ðŸ”¸ `nonlocal` â€“ modify variables from **enclosing (non-global)** scope

```python

    x = "outer"

        nonlocal x
        x = "modified outer"
        print("Inner:", x)

    inner()
    print("Outer:", x)

```

**Output:**

```

Outer: modified outer
```

---

| Namespace Type | Created                     | Destroyed                    | Contains                    | Example                 |
| -------------- | --------------------------- | ---------------------------- | --------------------------- | ----------------------- |
| **Built-in**   | Python starts               | When interpreter ends        | Built-ins (`print`, `len`)  | `print()`               |

| **Enclosing**  | Function definition (outer) | When outer function finishes | Variables in outer function | `outer()`               |

---

## ðŸ’¡ Key Takeaways

âœ… Every variable name in Python lives in some **namespace**.
âœ… Python resolves names using the **LEGB rule**.
âœ… Use `global` to modify global variables, `nonlocal` for enclosing ones.
âœ… Use `globals()` and `locals()` to inspect namespaces.
âœ… Each function call creates its **own local namespace**.
